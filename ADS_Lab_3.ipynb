```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("Crop_recommendation.csv")
```

```python
import matplotlib.pyplot as plt

# Scatter plot using Matplotlib
plt.scatter(df['N'], df['P'], c=df['rainfall'], cmap='viridis')
plt.xlabel('Nitrogen in soil')
plt.ylabel('Potassium in soil')
plt.title('Scatter Plot: Nitrogen vs Potassium with Rainfall')
plt.colorbar(label='Rainfall')
plt.show()

# Line plot using Matplotlib
plt.plot(df['ph'], label='pH of soil', marker='o')
plt.plot(df['temperature'], label='Temperature', marker='o')
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.title('Line Plot: pH of soil and Temperature')
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Scatter plot using Seaborn
sns.scatterplot(x='humidity', y='temperature', hue='rainfall', data=df, palette='viridis')
plt.title('Scatter Plot: Humidity vs Temperature with Rainfall')
plt.show()

# Pair plot using Seaborn
sns.pairplot(df, hue='rainfall', palette='viridis')
plt.title('Pair Plot: Pairwise relationships with Rainfall')
plt.show()
```

```python
import plotly.express as px

# 3D Scatter plot using Plotly
fig = px.scatter_3d(df, x='ph', y='temperature', z='humidity', color='rainfall')
fig.update_layout(title='3D Scatter Plot: pH vs Temperature vs Humidity with Rainfall')
fig.show()

# Line plot using Plotly
fig = px.line(df, x=df.index, y=['N', 'P'], markers=True)
fig.update_layout(title='Line Plot: Nitrogen and Potassium in soil')
fig.show()
```
